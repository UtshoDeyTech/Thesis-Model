# Counting Total Images

**The code defines two functions:**

- `count_images(directory)`: This function takes a directory path as input and returns the count of images `(files with extensions ".jpg", ".jpeg", or ".png")` present in that directory.
- `count_images_in_subdirs(parent_dir)`: This function takes a parent directory path as input and counts the total number of images present in all the subdirectories `(excluding any files that are not images)` under the parent directory. It calls the count_images function to count the number of images in each subdirectory and prints out the counts for each subdirectory. Finally, it returns the total count of images in all subdirectories.

Overall, the code can be used to count the number of images in a directory or in all subdirectories of a given parent directory.

```python
def count_images_in_subdirs(parent_dir):
    total_count = 0
    for sub_dir in os.listdir(parent_dir):
        sub_dir_path = os.path.join(parent_dir, sub_dir)
        if os.path.isdir(sub_dir_path):
            count = count_images(sub_dir_path)
            print(f"In directory '{sub_dir}', there are {count} image items.")
            total_count += count
    print(f"\nTotal number of Images in the '{parent_dir}' directory folder are {total_count} image items")
```

```python
def count_images(directory):
    count = 0
    for file in os.listdir(directory):
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            count += 1
    return count
```

**In this code, `parent_dir` is the path to the parent directory. The `os.listdir()` function is used to list the contents of the parent directory, which are then iterated through. The `os.path.join()` function is used to join the paths of the parent directory and subdirectories/files to create the full paths. The `os.path.isdir()` function is used to check if a given path is a directory, and the `os.path.isfile()` function is used to check if a given path is a file. The file names and subdirectory names are stored in separate lists, which are then used to create a Pandas DataFrame.**

```python
# Set the parent directory
parent_dir = "data"

# Initialize empty lists to store the file names and subdirectory names
file_names = []
sub_dir_names = []

# Iterate through the subdirectories of the parent directory
for sub_dir in os.listdir(parent_dir):
    sub_dir_path = os.path.join(parent_dir, sub_dir)
    
    # Check if the current subdirectory is a directory
    if os.path.isdir(sub_dir_path):
        # Iterate through the files in the subdirectory
        for file in os.listdir(sub_dir_path):
            file_path = os.path.join(sub_dir_path, file)
            
            # Check if the current file is a file (not a directory)
            if os.path.isfile(file_path):
                file_names.append(file)
                sub_dir_names.append(sub_dir)

# Create a Pandas DataFrame with the file names and subdirectory names
df = pd.DataFrame({
    'File Name': file_names,
    'Subdirectory Name': sub_dir_names
})
```

# Randomly Move 500 Images From Each Folder

**The script defines a function `move_random_images` that takes the source directory `src_dir` and the destination directory `dest_dir` as inputs, along with the number of images to be randomly chosen (default is 500). The function iterates over each sub-directory in `src_dir` and selects either `num_images` random images or all the images if there are less than num_images images in the `sub-directory`. The selected images are then cut to the sub-directory in the dest_dir with the same name as the original sub-directory.**


--------------------------------


**This function, move_random_images, takes in 3 arguments: src_dir, dest_dir, and num_images.**

- src_dir is the source directory where the subdirectories and images are stored.

- dest_dir is the destination directory where the selected random images will be moved to.

- num_images is the number of images to be selected randomly from each subdirectory. It has a default value of 500.

- The function iterates through each subdirectory in src_dir and selects num_images of randomly chosen images (if there are more images in the subdirectory than num_images). It then creates the same subdirectory structure in dest_dir and moves the selected random images from the source subdirectory to the destination subdirectory.

```python
def move_random_images(src_dir, dest_dir, num_images=750):
    for sub_dir in os.listdir(src_dir):
        sub_dir_path = os.path.join(src_dir, sub_dir)
        if os.path.isdir(sub_dir_path):
            image_files = [os.path.join(sub_dir_path, image_file) for image_file in os.listdir(sub_dir_path) if image_file.endswith('.jpg')]
            if len(image_files) > num_images:
                random_images = random.sample(image_files, num_images)
            else:
                random_images = image_files
            dest_sub_dir = os.path.join(dest_dir, sub_dir)
            if not os.path.exists(dest_sub_dir):
                os.makedirs(dest_sub_dir)
            for image_file in random_images:
                shutil.move(image_file, dest_sub_dir)
```

**The no_hair function:**

- Reads an image file located at `image_path` using `cv2.imread` and stores the result in a variable called `src`.

- Converts the `src` image from RGB color space to grayscale using `cv2.cvtColor`. The result is stored in a variable called `grayScale`.

- Creates a structuring element with a size of (17, 17) using `cv2.getStructuringElement`, and stores the result in a variable called `kernel`.

- Applies a black hat morphological operation on the `grayScale` image using `cv2.morphologyEx` with the `kernel` and stores the result in a variable called `blackhat`.

- Thresholds the `blackhat` image using `cv2.threshold`, where the threshold value is set to 10, and the desired type is binary. The result is stored in a tuple called `ret` and `thresh2`.

- Inpaints the `src` image using `cv2.inpaint` and the `thresh2` image as the mask. The algorithm used is "Telea". The result is stored in a variable called `dst`.

- Returns the final `dst` image.

```python
def no_hair(image_path):
    src = cv2.imread(image_path)
    grayScale = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
    kernel = cv2.getStructuringElement(1, (17, 17))
    blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)
    ret, thresh2 = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)
    dst = cv2.inpaint(src, thresh2, 1, cv2.INPAINT_TELEA)
    return dst
```

**This function "remove_skin_hair" is a python function that takes two parameters: "src_dir" and "dest_dir". The function performs the following operations:**

- Iterates over all the sub-directories in the source directory specified by the `"src_dir"` parameter.

- For each sub-directory, it creates a corresponding sub-directory in the destination directory specified by the `"dest_dir"` parameter, if it doesn't exist.

- For each image file ending with `.jpg`, `.jpeg` or `.png` in the sub-directory, it applies the `"no_hair"` function on the image, which should be a function that removes hair from the image.

- Saves the processed image with a modified file name (by appending "_modified" to the original name) in the corresponding sub-directory of the destination directory.

```python
def remove_skin_hair(src_dir, dest_dir):
    for sub_dir in os.listdir(src_dir):
        sub_dir_path = os.path.join(src_dir, sub_dir)
        if os.path.isdir(sub_dir_path):
            dest_sub_dir = os.path.join(dest_dir, sub_dir)
            if not os.path.exists(dest_sub_dir):
                os.makedirs(dest_sub_dir)
            for image_file in os.listdir(sub_dir_path):
                if image_file.endswith('.jpg') or image_file.endswith('.jpeg') or image_file.endswith('.png'):
                    image_path = os.path.join(sub_dir_path, image_file)
                    img_no_hair = no_hair(image_path)
                    dest_image_path = os.path.join(dest_sub_dir, image_file.replace('.jpg', '_modified.jpg'))
                    cv2.imwrite(dest_image_path, img_no_hair)
```
